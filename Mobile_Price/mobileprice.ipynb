```python
def removal_boxplot(df,column,threshold):
    sns.boxplot(y='Ratings',data=df)
    plt.title('Original')
    plt.show()

    removed_otliars=df[df[column]>=threshold]
    sns.boxplot(y=removed_otliars[column])
    plt.title('Removed')
    plt.show()
    return removed_otliars
threshold_value=3
no_outliers=removal_boxplot(df,'Ratings',threshold_value)
```

```python
sns.boxplot(y='RAM',data=df)
```

```python
Q1 = df['RAM'].quantile(0.25)
Q3 = df['RAM'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df=df[(df['RAM'] >= lower_bound) & (df['RAM'] <= upper_bound)]
sns.boxplot(data=df['RAM'])
plt.title("Boxplot without Outliers")
plt.show()
```

```python
#encoding
dfe=pd.get_dummies(df[['Brand']],drop_first=True,dtype=int)
dfe
```

```python
df1=pd.concat([df,dfe],axis=1)
df1
```

```python
df1.drop(['Brand me','Brand'],axis=1,inplace=True)
df1
```

```python
x=df1.drop(['Price'],axis=1)
x
```

```python
y=df['Price']
y
```

```python
#train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
#model creation
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(criterion='absolute_error')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred

```

```python
import joblib
joblib.dump(model,'model.pkl')
joblib.dump(x.columns.tolist(), 'feature_names.pkl')
```

```python
#DataFrame
df3=pd.DataFrame({'Actual_Value':y_test,'Predicted_Value':y_pred})
df3
```

```python
#perfomance
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
print('mean_squared_error: ',mean_squared_error(y_test,y_pred))
print('root_mean_squared_error: ',root_mean_squared_error(y_test,y_pred))
print('Mean absolute percentage error:  ',mean_absolute_percentage_error(y_test,y_pred))
print('r2 score: ',r2_score(y_test,y_pred))
```
